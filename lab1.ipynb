```python
# 日期確認
def is_date_valid(d):
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100

    if year >= 1:
        if 12 >= month >= 1:
            month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            days_max = month_days[month-1]
            if is_leap_year(year) and month == 2:
                days_max = days_max + 1
            if days_max >= day >= 1:
                return True
            else:
                return False
        else:
            return False
    else:
        return False

# print(is_date_valid(200010229))
```

# 20260507 群集型態: list/dict/**set**/tuple

list: 同類型的清單 [學生, 學生, 學生], list的key是固定的 0 1 2 3 (有序)

dict: 聚集多個欄位(key:value)來成為複雜單位 {"name":"Elwing", "身高":175} (無序)

**set: 同類型但不重複的集合(無序)** {30, 20, 15, 30} 沒有key

**tuple: 簡化版本的字典** {"r":255, "g":128, "b":200}(255, 128, 200) key=0 1 2 操作: tuple[key]

```python
a = {30, 2, 1}
b = {30, 40, 50}

# 交集(&): A B 兩集合都有的
print(a & b)

# 聯集(|): 把 A B 有的統整起來
print(a | b)

# 減集(-): A - B 把 A 的東西扣掉 B 有的(A - AB交集)
print(a - b)

# 迴圈應用

result = []
# 設定空的群集方式:
# set()
# list()
# dict()

for i in (a | b):
    # print('!', i)
    result.append(i)
print(result)
```

# -- set() 練習 - 樂透抽號碼

```python
import random
count = 6
lotto = set()

while True:
    n = random.randint(1, 10)
    lotto.add(n)
    # 重複的號碼時繼續抽
    # set長度滿的時候才結束迴圈
    if len(lotto) == count:
        break;

print(lotto)
```

# 20260509 踩地雷應用: 找出最小值出現在哪個位置

```python
l = [80, 40, 50, 30, 20]

# 找出最小值 & 最小值出現在哪個位置
# 設定最大值: float("inf")
# 設定最小值: float("-inf")
minv = float("inf")
mini = float("-inf")
for i in range(len(l)):
    n = l[i]
    if n < minv:
        minv = n
        mini = i
    print(minv, mini)
```

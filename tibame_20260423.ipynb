# 完全平方日

當天日期 (20241001) 開根號是整數 (即 $4499^2$)

題目: 從20241001往後數出10個完全平方日

確認1. 數值判斷是否為日期

確認2. 不是閏年(每4年1閏, 每100年不閏, 每400年閏)

```python
# 確認2. 判斷Year 是否閏年
# boolean
def is_leap(self):
    if self % 400 == 0:
        return True
    elif self % 100 == 0:
        return False
    elif self % 4 == 0:
        return True
    else:
        return False

class Date:
    # 設定Date(年,月,日)
    def set_data(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    # 確認是否符合日期格式
    def is_date(self):
        if (31 >= day > 0) and (12 >= month > 0) and (year >= 1911):
            return True
        else:
            return False

i = 4499
count = 0

while count < 10:

    d = i * i           #數值d = 20241001拆成年月日
    day = d % 100       #日:   01, 20241001除100取餘數
    d_tmp = d // 100    #除100取商, 替換d = 202410, 用來取月
    month = d_tmp % 100 #月:   10, 202410除100取餘數
    year = d_tmp // 100 #年: 2024, 202410除100取商

    # 設定: 取年year 月month 日day
    date = Date()
    date.set_data(year, month, day)

    # 排除閏年 and 確認為日期樣式: 年(1911-), 月(1-12), 日(1-31)
    if not is_leap(year) and date.is_date():
        print('完全平方日:', d, '[', year, '年', month, '月', day,'日 ]')
        count += 1
    i = i + 1
```
